```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 4, 5])
```

```python
plt.scatter(x, y)
```

```python
def mean(arr):
    return(sum(arr) / len(arr))
```

```python
m = 0

# mean of both the arrays
x_mean = mean(x)
y_mean = mean(y)

numerator = 0
denominator = 0


for i in range(len(x)):
    numerator += (x[i]-x_mean) * (y[i] - y_mean)
    denominator += pow((x[i] - x_mean), 2)

m = numerator / denominator

print(m)
```

```python
c = y_mean - m*x_mean
print(c)
```

```python
x_line = np.linspace(0, 5, 100)
y_line = m*x_line + c
plt.plot(x_line, y_line, '-r', label='y=2x+1')
plt.scatter(x, y)
plt.title('Graph of y=2x+1')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
iter = 1000
theta = np.random.rand()
b = np.random.rand()
y_pred = theta * x + b
alpha = 0.001
thetaa = list()
n = x.shape[0]
cost_list = []

for i in range(1, iter+1):

    y_pred = theta*x + b
    cost = sum(((y_pred - y)**2)) / (2*n)
    cost_list.append(cost)
    theta = theta - (sum((y_pred - y) *x) * alpha) / (n)
    b = b - (sum((y_pred - y) *x) * alpha) / (n)

    print(f"Iteration: {i}\tCost: {cost}")
    if i %200 == 0:
        thetaa.append(theta)
```
