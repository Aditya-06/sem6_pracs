```python
def encrypt(text,k):

    ct=''
    for i in range(len(text)):
        char=text[i]   #getting one plain text character at a time starting from index position 0 upto the end
        if (char.isupper()):
            x=chr((ord(char)-65+k)%26+65)
            ct+=x
            
        if (char.islower()):
            x=chr((ord(char)-97+k)%26+97)
            ct+=x
            
        if (ord(char)==32):
            ct+='X'
            
    return ct
        
def decrypt(text,k):
    ot=''
    for i in range(len(text)):
        char=text[i]
            
        if(char.isupper()):
            x=chr((ord(char)-65-k)%26+65)
            ot+=x
                       
        if(char.islower()):
            x=chr((ord(char)-97-k)%26+97)
            ot+=x
        
    print('Original text was',ot)
    
while(True):
    print('Caesar Cipher Encryption and Decryption')
    print('\n1.Encryption\n2.Decryption\n3.Exit')
    choice=int(input('Choose your operation:'))
    if (choice==1):
        pt=input("Enter plain text:")
        key=int(input("Enter key value:"))
        print('Encrypted text is:',encrypt(pt,key))
    elif (choice==2):
        ct=encrypt(pt,key)
        decrypt(ct,key)
    elif (choice==3):
        break
```

```python
def encrypt(text,k):
    
    alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'

    ct=''
    for i in range(len(text)):
        char=text[i]   #getting one plain text character at a time starting from index position 0 upto the end
        char_index=alphabet.find(char)
        cipher_index=(char_index+k)%36
        ct+=alphabet[cipher_index]
            
    return ct
        
def decrypt(text,k):
    
    alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    
    ot=''
    for i in range(len(text)):
        cipher_value=text[i]
        cipher_index=alphabet.find(cipher_value)
        char_index=(cipher_index-k)%36
        ot+=alphabet[char_index]
        
    return ot

def brute_force(text):
    
    alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    
    for j in range(len(alphabet)):
        k=j
        ot=''
        for i in range(len(text)):
            cipher_value=text[i]
            cipher_index=alphabet.find(cipher_value)
            char_index=(cipher_index-k)%36
            ot+=alphabet[char_index]
        print('Original text for k{} = {}'.format(j,ot))
    
while(True):
    print('Caesar Cipher Encryption and Decryption')
    print('\n1.Encryption\n2.Decryption\n3.Brute-Force Attack\n4.Exit')
    choice=int(input('Choose your operation:'))
    if (choice==1):
        pt=input("Enter plain text (in caps and no spaces, digits allowed):")
        key=int(input("Enter key value:"))
        print('Encrypted text is:',encrypt(pt,key))
    elif (choice==2):
        ct=input("Enter cipher text (in caps and no spaces, digits allowed):")
        key=int(input("Enter key value:"))
        print('Original text is:',decrypt(ct,key))
    elif (choice==3):
        ct=input("Enter cipher text (in caps and no spaces, digits allowed):")
        brute_force(ct)
    elif (choice==4):
        break
```
