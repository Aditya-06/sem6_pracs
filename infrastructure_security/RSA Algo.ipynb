```python
def gcd(x,y):
    if(y==0):
        return x
    else:
        return gcd(y, x%y)
    
def get_d(phi_n,e):
    d=None
    for i in range(e*25):
        if((i*e)%phi_n==1):
            d=i
            
    return d

def encryption_keys(p,q):
    
    n=p*q
    phi_n=(p-1)*(q-1)
    
    E=int(input('Enter value for e:'))
    if(gcd(phi_n,E)==1):
        e=E
    else: 
        print('Try again')
    
    d=get_d(phi_n,e)
    return e,d
    
def decrypt(lst,d,n):

    alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    
    temp=[]
    ot=''
    
    for i in range(len(lst)):
        temp.append((lst[i]**d)%n)
        
    for i in range(len(temp)):
        ot+=alphabet[temp[i]]
        
    return ot
    
def encrypt(mes):
    
    P=int(input("Enter value of p (prime):"))
    Q=int(input("Enter value of q (prime):"))
    n=P*Q

    e,d=encryption_keys(P,Q)
    
    ct=[]
    
    alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    
    temp=[]
    for char in mes:
        char_index=alphabet.find(char)
        temp.append(char_index)
        
    print(temp)
    for i in range(len(temp)):
        ct.append((temp[i]**e)%n)
        
    print('Public key =',(e,n))
    print('Private key =',(d,n))
    print("Cipher text:",ct)
    print("Original text was:",decrypt(ct,d,n))

print("RSA Algorithm")
while(True):
    pt=input("Enter plain text:")
    encrypt(pt)
    break 
```
